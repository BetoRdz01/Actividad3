```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('Ventas_totales.csv') 
data
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data2=data.copy()
```

```python
#Sustituir valores nulos con promedio
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data2["tarjetas_debito"].mean(),1))
data2
```

```python
#Sustituir valores nulos con promedio
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data2["salon_ventas"].mean(),1))
data2
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data2["tarjetas_credito"] =data2["tarjetas_credito"].fillna(method="ffill") 
data2
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data2["otros_medios"] =data2["otros_medios"].fillna(method="ffill") 
data2
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data2["subtotal_ventas_alimentos_bebidas"] =data2["subtotal_ventas_alimentos_bebidas"].fillna(method="bfill") 
data2
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data2["almacen"] =data2["almacen"].fillna(method="bfill") 
data2
```

```python
#Sustituir valores nulos con mediana
data2["otros"]=data2["otros"].fillna(round(data2["otros"].median(),1))
data2
```

```python
#Sustituir valores nulos con mediana
data2["electronicos_articulos_hogar"]=data2["electronicos_articulos_hogar"].fillna(round(data2["electronicos_articulos_hogar"].median(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Eliminamos filas con nulos
data3= data2.dropna(0)
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#####Empezamos con la eliminación de outliers
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data3["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Corrientes")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data3["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data3[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data4= data3[(y<=Limite_Superior)&(y>=Limite_Inferior)]
```
